I've also found it effective to fetch results for multiple items for comparison, here, I am fetching results for all hardcoded item recipes

```python
# creating a list of all item ids for the items that have hardcoded recipes
cores = list(Crafting_Item.HARDCODED_RECIPES.keys())

return_dict = {}

# looping through all the cores and getting the analysis metrics for each, storing them in a dictionary

for core in cores:
    item = Crafting_Item.CraftingItem(core)

    return_dict[core] = item.get_analysis_metrics()
```

```python
# Creating a dataframe from the dictionary and printing out the results here
df = pd.DataFrame.from_dict(return_dict, orient='index')

df.head(len(df))
```

```python
# Sorting the dataframe by profit per instant sale scaled by amount of raw ingredients needed on the Trading Post and printing the top 10 results
df.sort_values(by='profit_per_raw_insta', ascending=False).head(10)
```

Creating a simple visualization to inform what item would be the best to craft, comparing total profit margin to profit per raw resource, adding volume as a scaling metric to get an idea of the size of the market for the item in question. Here we can see that Fine Rift Motivations might be a viable item, despite their smaller volume. Tier 3 cores also emerge as a potential option as long as they are sold en masse to compensate for their low total profitability.

```python
# Set style
sns.set_theme(style="whitegrid")

# Price vs. Crafting Cost table visualizing profitablity of items vs amount of work required to craft them
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['profit_per_raw'], y=df['profit_margin'], hue=df['item_name'], size=df['volume'], sizes=(20,200), palette='tab10', s=100)
plt.title("Total Profit Margin vs. Profit Per Raw Resource")
plt.xlabel("Profit Per Raw Resource")
plt.ylabel("Total Profit Margin")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
