```python
# import required modules

import Crafting_Item
import pandas as pd
```

Below is an example of a single crafting item and how to effectively use the object to get trading post metrics and raw ingredients for the item Jade Bot Core: Tier 10

```python
Sample_Item = Crafting_Item.CraftingItem(96613) # Bringing in Jade Bot Core: Tier 10

Sample_Item.get_analysis_metrics() # Getting the analysis metrics for the item
```

```python
print(Sample_Item.metrics_dict) # Printing the dictionary to the console

print(Sample_Item.raw_ingredients_dataframe) # Printing the raw ingredients dataframe to the console
```

```python
# I've also found it effective to fetch results for multiple items for comparison, here, I am fetching results for all hardcoded item recipes
cores = list(Crafting_Item.HARDCODED_RECIPES.keys())

return_dict = {}

for core in cores:
    item = Crafting_Item.CraftingItem(core)

    return_dict[core] = item.get_analysis_metrics()
```

```python
# Creating a dataframe from the dictionary and printing out the results here
df = pd.DataFrame.from_dict(return_dict, orient='index')

df.head(len(df))
```

```python
df.sort_values(by='profit_per_raw_insta', ascending=False).head(10) # Sorting the dataframe by profit per instant sale on the Trading Post and printing the top 10 results
```
